```python

def Store_Stock():
    
    
    # User Define function for taking user input to get the number of item
    def f_Biscuits():
        Biscuits=int(input("How many buiscuit packets are required: \n"))    
        return Biscuits
    
    def f_Rice():
        Rice=int(input("How many Rice packets are required: \n"))    
        return Rice
    
    def f_Pulses():
        Pulses=int(input("How many Pulses packets are required: \n"))    
        return Pulses
    
    def f_Spices():
        Spices=int(input("How many Spices packets are required: \n"))    
        return Spices
    
    def f_Choclates():
        Choclates=int(input("How many Choclates are required: \n"))    
        return Choclates
    
    
        #Logical functions with conditions
    def stock_c1(): 
        Biscuits=f_Biscuits()
        if stock[0]<Biscuits:
            print("Out of stock")            
            buy.insert(0,0)                
        else:
            
            
            buy.insert(0,Biscuits)            
    stock_c1()
    
    
    def stock_c2(): 
        Rice=f_Rice()
        if stock[1]<Rice:
            print("Out of stock")            
            buy.insert(1,0)                
        else:
           
            buy.insert(1,Rice)        
    stock_c2()    
    
        
    def stock_c3(): 
        Pulses=f_Pulses()
        if stock[2]<Pulses:
            print("Out of stock")            
            buy.insert(2,0)                
        else:
            
            buy.insert(2,Pulses) 
    stock_c3()
        
                
    def stock_c4(): 
        Spices=f_Spices()
        if stock[3]<Spices:
            print("Out of stock")            
            buy.insert(3,0)                
        else:
            
            buy.insert(3,Spices)
    stock_c4()        

        
    def stock_c5(): 
        Choclates=f_Choclates()
        if stock[4]<Choclates:
            print("Out of stock")            
            buy.insert(4,0)                
        else:
            
            buy.insert(4,Choclates)
    stock_c5()
    return


```

```python
def Store_invoice():        
    #printing Invoice
        
    print("---------------------------------Grocery Store---------------------------------\n")
    print("**********************************TAX Invoice**********************************\n")
    
    print("-------------------------------------------------------------------------------")
    print("%10s %10s %10s %10s %10s %s %10s %10s" %("S.No","Items","quantity","Cost/Item","Tax", "  ","Tax Amount","Total Cost"   ))
    print("-------------------------------------------------------------------------------")
    print("%10d %10s %10d %10d %10d %s %10d %10d" %(1,"Biscuits" ,buy[0], cost[0],   tax[0],"%" , (buy[0]*10*tax[0])/100, (buy[0]*10)+(buy[0]*10*tax[0])/100))
    print("%10d %10s %10d %10d %10d %s %10d %10d" %(2,"Rice"     ,buy[1], cost[1],  tax[1],"%" , (buy[1]*10*tax[1])/100, (buy[1]*50)+(buy[1]*10*tax[1])/100))
    print("%10d %10s %10d %10d %10d %s %10d %10d" %(3,"Pulses"   ,buy[2], cost[2],  tax[2],"%" , (buy[2]*10*tax[2])/100, (buy[2]*120)+(buy[2]*10*tax[2])/100))
    print("%10d %10s %10d %10d %10d %s %10d %10d" %(4,"Spices"   ,buy[3], cost[3],  tax[3],"%" , (buy[3]*10*tax[3])/100, (buy[3]*200)+(buy[3]*10*tax[3])/100))
    print("%10d %10s %10d %10d %10d %s %10d %10d" %(5,"Choclates",buy[4], cost[4],  tax[4],"%" , (buy[4]*10*tax[4])/100, (buy[4]*1)+(buy[4]*10*tax[4])/100))
    print("-------------------------------------------------------------------------------")
    
    Total_item_Cost = (buy[0]*10)+(buy[1]*50)+(buy[2]*120)+(buy[3]*200 )+(buy[4]*1)
    print("Total Item Cost     :",Total_item_Cost,"$" )
    print("Delivery charges    :",Delivery_charges,"$")
    print("                     ----------")
    TotalBill = Total_item_Cost + int(Delivery_charges)
                                    
    print("Total Bill Amount   :",TotalBill,"$\n\n")
    print("***************************************************************************")
                                    
    return TotalBill
    
    
    
```
